```python
# Let's create a function for adjusted r2 for test data set
def adj1_r2(x,y):
    r2 = Lr.score(x_test, y_test)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = (1-(1-r2)*(n-1)/(n-p-1))
    return adjusted_r2
```
