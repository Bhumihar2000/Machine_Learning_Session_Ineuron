# Step-2 => We can try to find out the covariance matrix

# Step-3 => then we can try to find out the eigan valve and eigan vector.

# But here we are using library first.


```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(df)
```

```python
pd.DataFrame(pca.fit_transform(df))
```

```python
pca.explained_variance_ratio_
```

# Screen Plot => That represent the relationship between Explained variance ratio and number of principal component that is required.

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("number of principal component required")
plt.ylabel("Explained Variance Ratio")
plt.show()
```

```python
pca1 = PCA(n_components=5)
pca1.fit_transform(df)
new_data = pd.DataFrame(pca1.fit_transform(df), columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
new_data
```
